```python
#12. Afficher les directeurs qui ont produit le plus de films/séries disponibles sur Netflix
# lire le fichier csv
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
donnees=pd.read_csv("/home/souad/data-dev/correction-exo/python/PROJET-NETFLIX/netflix_titles.csv", index_col='show_id')
donnees.head()
```

```python
#12. Afficher les directeurs qui ont produit le plus de films/séries disponibles sur Netflix
directors=pd.Series(", ".join(donnees["director"].dropna()).split(", "))

directors.value_counts().head(5)
```

```python
donnees['director'].value_counts()
```

```python
#13. Voir si Jan Suter travaille souvent avec les mêmes acteurs

directors_notna=donnees[donnees['director'].notna()]
donnees_jan_suter=directors_notna[directors_notna['director'].str.contains("Jan Suter")]    
donnees_jan_suter
jan_suter_cast=pd.Series(', '.join(donnees_jan_suter['cast'].dropna()).split(', '))
jan_suter_cast.value_counts().head(5)
```

```python
#14. Représenter les dix pays qui ont produit le plus de conte-nus disponibles sur Netflix, avec le nombre de contenus parpays
donnees_pays=pd.Series(", ".join(donnees['country'].dropna()).split(", "))
donnees_pays=donnees_pays.value_counts().head(10)
donnees_pays
```

```python
#suite 14
#donnees.loc[donnees['country'].isin(donnees_pays.index)]
dataframe_pays=donnees[donnees["country"].isin(donnees_pays.index)]
dataframe_pays
```

```python
#Tracer un countplot
plt.figure(figsize=[14,6])
sns.countplot(data=dataframe_pays, x='country')
plt.xticks(rotation=90)
plt.show()
```

```python
#15 Tracer un graphe à barres du nombre de films/séries par classement de contenu (rating)
plt.figure(figsize=[14,6])
sns.countplot(data=donnees, x='rating', order=donnees['rating'].value_counts().keys())
plt.xticks(rotation=90)
plt.show()
```

```python

#16. Afficher l’évolution du nombre de films/séries disponibles sur Netflix au cours du temps
data_inter=donnees[donnees['date_added'].notna()]
dates=pd.to_datetime(data_inter['date_added']).dt.year
print(dates)

```

```python
#suite 16
#ajout la colonne year_added au dataframe donnees

donnees['year_added']=dates
donnees
```

```python
#regrouper les donnees 
donnees_grp=donnees.groupby(['type','year_added']).size().reset_index(name='Count')
donnees_grp
```

```python
#courbe evolution du nombre de films au fil des années
sns.pointplot('year_added', 'Count',hue='type' , data=donnees_grp)

plt.xticks(rotation=90)
plt.show()
```

```python
#17 Afficher la distribution de la durée des films disponiblessur Netflix
donnees_movie=donnees.loc[donnees['type']=='Movie']
duree=pd.Series(donnees_movie['duration']).str.replace("min","")
duree
```

```python
#histogramme durée des films
sns.distplot(duree)
```
