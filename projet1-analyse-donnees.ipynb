```python
#Charger le fichier « pole_emploi.xlsx » après l’avoir ouvert et visualiser et l’assigner à la variable « p_emploi ».
p_emploi=pd.read_excel('//home//souad//support-cours//correction-exo//pythnon//projet1-analyse-donnees//pole_emploi.xlsx', sheet_name= 'Data', skiprows=3)
print(p_emploi)
```

```python
#Joindre les data et p_emploi au seins d’un seul et unique dataframe en faisant correspondre les régions par leur codes. Utiliser pd.merge() pour réaliser cette jointure.

data7=pd.merge(data,p_emploi, how='inner', left_on='REG', right_on='Code');
data7
```

```python
#Créer une colonne quantile dans p_emploi qui prendra les valeurs [‘q25’, ‘q50’, ‘q75’, ‘q100’] en fonction du quartile auquel appartient le nombre de pôle emploi. Pour se faireutiliser « pd.qcut() ».
data7['Quantile']=pd.qcut(data7['Pôle emploi 2018'], 4, labels=['q25 ', 'q50', 'q75', 'q100'])
data7
```

```python
#Générer un pivot de table montrant la moyenne du taux de chômage du trimestre 1 de l’année 2020 pour chaque régions et quantile du nombre de pôle emploi.
table1= pd.pivot_table(data7, index=data['LIBREG'], columns='Quantile' , values='2020-T1')
table1
```
