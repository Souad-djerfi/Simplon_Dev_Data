```python
import pandas as pd 

data=pd.read_excel('//home//souad//data-dev//correction-exo//pythnon//projet1-analyse-donnees//chomage_zone.xls',  sheet_name='txcho_ze', skiprows=5)
data
```

```python
#Calculer le taux de chômage moyen par régions pour chaque trimestres.
data_grp=data.groupby(data['LIBREG'], as_index=False).mean()
data_grp
```

```python
# Tracer diagramme de bar du taux de chomage du 1er trimestre 2020
data_grp.plot(x='LIBREG', y='2020-T1', kind='bar',  title='DIAGRAMME DE BARRE', grid=True, fontsize=12, figsize=(15,12))
```

```python
# isoler  les colonnes 3 à 73. et assignées les à data2.
data2=data.iloc[:,3:73]
data2
```

```python
#Calculer le taux de chômage moyen par régions pour chaque trimestres.
data_grp2= data2.groupby('LIBREG').mean()
data_grp2
```

```python
#Transposer le dataframe
data_grp2_trans=data_grp2.transpose()
data_grp2_trans
```

```python
# reindexer le dataframe
data_grp2_trans=data_grp2_trans.reset_index()
data_grp2_trans
```

```python
# renommer la colonne index en periode
data_grp2_transR=data_grp2_trans.rename(columns={'index':'periode'})
data_grp2_transR
```

```python
# tracer la ligne représentant l’évolution du taux de chômage pour la région ‘Provence-Alpes-Côte d'Azur’.

data_grp2_transR.plot(x='periode', y="Provence-Alpes-Côte d'Azur", figsize=(15,12))
```

```python
#Répéter l’étape précédente sans préciser l’argument « y = ».
data_grp2_transR.plot(x='periode', figsize=(15,12))
```

```python
#Représenter l’évolution du taux de chômage pour la région ‘Provence-Alpes-Côte d'Azur’, ‘Occitanie’ et ‘Ile-de-France’ sur un même graphique.
data_grp2_transR.plot(x='periode', y={"Provence-Alpes-Côte d'Azur",'Occitanie', 'Ile-de-France'}, figsize=(15,12))
```
